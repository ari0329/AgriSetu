```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error

# -----------------------------
# 1. Load Dataset
# -----------------------------
data = pd.read_excel("smart_agriculture_ml_dataset_50000.xlsx")


```

```python
# -----------------------------
# 2. Feature Selection
# -----------------------------
X = data[[
    "Soil_Moisture_%",
    "Soil_Temperature_C",
    "Rainfall_ml",
    "Air_Temperature_C",
    "Humidity_%"
]]

# Targets
y_crop = data["Recommended_Crop"]
y_months = data["Growth_Duration_Months"]

```

```python
# -----------------------------
# 3. Encode Crop Labels
# -----------------------------
label_encoder = LabelEncoder()
y_crop_encoded = label_encoder.fit_transform(y_crop)

# -----------------------------
# 4. Train-Test Split
# -----------------------------
X_train, X_test, y_crop_train, y_crop_test = train_test_split(
    X, y_crop_encoded, test_size=0.2, random_state=42
)

_, _, y_months_train, y_months_test = train_test_split(
    X, y_months, test_size=0.2, random_state=42
)

```

```python
# -----------------------------
# 5. Train Models
# -----------------------------
# Crop Prediction (Classification)
crop_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
crop_model.fit(X_train, y_crop_train)

# Growth Duration (Regression)
month_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
month_model.fit(X_train, y_months_train)

```

```python
# -----------------------------
# 6. Predictions
# -----------------------------
crop_pred = crop_model.predict(X_test)
month_pred = month_model.predict(X_test)

# -----------------------------
# 7. Evaluation
# -----------------------------
crop_accuracy = accuracy_score(y_crop_test, crop_pred)
mae = mean_absolute_error(y_months_test, month_pred)
rmse = np.sqrt(mean_squared_error(y_months_test, month_pred))

print("Crop Prediction Accuracy:", crop_accuracy * 100, "%")
print("Growth Duration MAE:", mae)
print("Growth Duration RMSE:", rmse)

```

```python
import pandas as pd
from datetime import datetime
from reportlab.platypus import (
    SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors

# -----------------------------
# 1. Sample New Sensor Input
# -----------------------------
sample_input = pd.DataFrame([{
    "Soil_Moisture_%": 1026,
    "Soil_Temperature_C": 33,
    "Rainfall_ml": 81,
    "Air_Temperature_C": 24,
    "Humidity_%": 90
}])

# -----------------------------
# 2. Model Predictions
# -----------------------------
pred_crop = crop_model.predict(sample_input)
pred_month = month_model.predict(sample_input)

recommended_crop = label_encoder.inverse_transform(pred_crop)[0]
growth_months = round(pred_month[0])

print("Recommended Crop:", recommended_crop)
print("Growth Duration (Months):", growth_months)

#------------------------------
#save the model 
#------------------------------
import joblib

joblib.dump(crop_model, "crop_model.pkl")
joblib.dump(month_model, "month_model.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")

print("Models Saved Successfully")

# -----------------------------
# 3. PDF Setup
# -----------------------------
timestamp = datetime.now()
date_time_str = timestamp.strftime("%d_%m_%Y__%H_%M_%S")
display_time = timestamp.strftime("%d %b %Y | %H:%M:%S")

file_name = f"AgriSetu_{date_time_str}.pdf"

pdf = SimpleDocTemplate(
    file_name,
    pagesize=A4,
    rightMargin=40,
    leftMargin=40,
    topMargin=40,
    bottomMargin=40
)

styles = getSampleStyleSheet()

# -----------------------------
# 4. Custom Styles
# -----------------------------
title_style = ParagraphStyle(
    "TitleStyle",
    parent=styles["Title"],
    alignment=1,
    textColor=colors.white,
    fontSize=20
)

section_style = ParagraphStyle(
    "SectionStyle",
    parent=styles["Heading2"],
    textColor=colors.HexColor("#2E7D32"),
    spaceBefore=16,
    spaceAfter=8
)

normal_style = ParagraphStyle(
    "NormalStyle",
    parent=styles["Normal"],
    fontSize=11,
    leading=15,
    spaceAfter=10
)

content = []

# -----------------------------
# 5. Header Banner
# -----------------------------
header_table = Table(
    [[Paragraph("SMART CROP PREDICTION REPORT â€“ AGRISETU", title_style)]],
    colWidths=[450]
)
header_table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, -1), colors.HexColor("#2E7D32")),
    ("ALIGN", (0, 0), (-1, -1), "CENTER"),
    ("VALIGN", (0, 0), (-1, -1), "MIDDLE"),
    ("TOPPADDING", (0, 0), (-1, -1), 14),
    ("BOTTOMPADDING", (0, 0), (-1, -1), 14),
]))

content.append(header_table)
content.append(Spacer(1, 20))
content.append(Paragraph(f"<b>Generated On:</b> {display_time}", normal_style))

# -----------------------------
# 6. Sensor Data Section
# -----------------------------
content.append(Paragraph("INPUT SENSOR DATA", section_style))

sensor_data = [["Parameter", "Value"]]
for col, val in sample_input.iloc[0].items():
    sensor_data.append([col.replace("_", " "), str(val)])

sensor_table = Table(sensor_data, colWidths=[260, 160])
sensor_table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, 0), colors.HexColor("#A5D6A7")),
    ("GRID", (0, 0), (-1, -1), 0.5, colors.grey),
    ("FONT", (0, 0), (-1, 0), "Helvetica-Bold"),
    ("ALIGN", (1, 1), (-1, -1), "CENTER"),
    ("BOTTOMPADDING", (0, 0), (-1, -1), 8),
    ("TOPPADDING", (0, 0), (-1, -1), 8),
]))

content.append(sensor_table)

# -----------------------------
# 7. Prediction Result (TABLE)
# -----------------------------
content.append(Spacer(1, 18))
content.append(Paragraph("PREDICTION RESULT", section_style))

result_table = Table([
    ["Recommended Crop", recommended_crop],
    ["Expected Growth Duration", f"{growth_months} Months"]
], colWidths=[260, 160])

result_table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, -1), colors.HexColor("#E8F5E9")),
    ("GRID", (0, 0), (-1, -1), 0.5, colors.green),
    ("FONT", (0, 0), (0, -1), "Helvetica-Bold"),
    ("ALIGN", (1, 0), (-1, -1), "CENTER"),
    ("TOPPADDING", (0, 0), (-1, -1), 10),
    ("BOTTOMPADDING", (0, 0), (-1, -1), 10),
]))

content.append(result_table)

# -----------------------------
# 8. Prediction Analysis (PARAGRAPH)
# -----------------------------
content.append(Spacer(1, 14))

prediction_paragraph = (
    f"Based on the analysis of real-time environmental sensor data, the machine learning model "
    f"recommends <b>{recommended_crop}</b> as the most suitable crop for cultivation under the "
    f"current soil and climatic conditions. The predicted growth duration of approximately "
    f"<b>{growth_months} months</b> enables effective crop planning, optimized use of resources, "
    f"and improved yield potential. This data-driven recommendation assists farmers in making "
    f"informed agricultural decisions while reducing uncertainty and risk."
)

content.append(Paragraph(prediction_paragraph, normal_style))

# -----------------------------
# 9. Footer Note
# -----------------------------
content.append(Spacer(1, 20))
content.append(Paragraph(
    "This report is generated by the AgriSetu Smart Agriculture System using machine learning "
    "models integrated with real-time sensor data.",
    styles["Italic"]
))

# -----------------------------
# 10. Build PDF
# -----------------------------
pdf.build(content)

print(f"\nâś… PDF Report successfully saved as '{file_name}'")
```
